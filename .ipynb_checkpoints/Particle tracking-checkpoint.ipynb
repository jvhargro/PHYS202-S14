```python
from Blobfinder import *
from PIL import Image
import numpy as np
```

```python
def Tracker(P, tau, deltap):
   
    distances = []


    for i in range(9):
        im1 = Image.open('run3/frame0000'+str(i)+'.jpg')
        im2 = Image.open('run3/frame0000'+str(i+1)+'.jpg')
        lst1 = BlobFinder(im1, 600)
        Bead1 = getBeads(P, lst1)
        lst2 = BlobFinder(im2, 600)
        Bead2 = getBeads(P, lst2)
        for blob in range(len(Bead1)):
            if blob < len(Bead2):
                distance = Bead1[blob].distanceTo(Bead2[blob])
                if distance <= deltap:
                    distances.append(distance)
    
    for i in range(10,99):
        im1 = Image.open('run3/frame000'+str(i)+'.jpg')
        im2 = Image.open('run3/frame000'+str(i+1)+'.jpg')
        lst1 = BlobFinder(im1, 600)
        Bead1 = getBeads(P, lst1)
        lst2 = BlobFinder(im2, 600)
        Bead2 = getBeads(P, lst2)
        for blob in range(len(Bead1)):
            if blob < len(Bead2):
                distance = Bead1[blob].distanceTo(Bead2[blob])
                if distance <= deltap:
                    distances.append(distance)
    
    for i in range(100,199):
        im1 = Image.open('run3/frame00'+str(i)+'.jpg')
        im2 = Image.open('run3/frame00'+str(i+1)+'.jpg')
        lst1 = BlobFinder(im1, 600)
        Bead1 = getBeads(P, lst1)
        lst2 = BlobFinder(im2, 600)
        Bead2 = getBeads(P, lst2)
        for blob in range(len(Bead1)):
            if blob < len(Bead2):
                distance = Bead1[blob].distanceTo(Bead2[blob])
                if distance <= deltap:
                    distances.append(distance)
    
    return distances
    
```

```python
distances = Tracker(25, 600, 25)
```

```python
np.savetxt('RadialDistances.txt', distances, fmt='%.4f')
!cat 'RadialDistances.txt'
```
