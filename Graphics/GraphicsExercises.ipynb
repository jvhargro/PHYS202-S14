```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
offsets = np.linspace(0, 2*np.pi, 4, endpoint=False)
# Create array with shifted-sine curve along each column
yy = np.transpose([np.sin(x + phi) for phi in offsets])

plt.rc('lines', linewidth=4)
fig, (ax0, ax1)  = plt.subplots(nrows=2)

plt.rc('axes', color_cycle=['r', 'b', 'y', 'g'])
ax0.plot(yy)
ax0.set_title('Color Cycle rbyg')

ax1.set_color_cycle(['c', 'm', 'y', 'k'])
ax1.plot(yy)
ax1.set_title('Color Cycle cmyk')

# Tweak spacing between subplots to prevent labels from overlapping
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
x = np.arange(-6,6, .255)
y = np.arange(-6,6, .255)
x, y = np.meshgrid(x, y)
G = np.exp(-x**2 - y**2)
surf = ax.plot_surface(x, y, G, rstride=1, cstride=1, cmap=cm.jet,
        linewidth=0, antialiased=False)
ax.set_zlim3d(-1.0, 1.0)
```

q1 = -1.0 nC at (-2 m, 0 m)
q2 = -1.0 nC at (+2 m, 0 m)
q3 = +1.0 nC at (0 m, -2 m)
q4 = +1.0 nC at (0 m, +2 m)

(a) Make a surface plot on the range -4 <(x,y)< 4 using 101 bins in x and 101 bins in y. In this case, do not let k = q = d = 1. Use dimensionful quantities and compute the value of the potential in volts.

Recall k = 8.9875517873681764 × 109 N⋅ m2/C2.


```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
x = np.arange(-4, 4, .079)
y = np.arange(-4, 4, .079)
x, y = np.meshgrid(x, y)
d = 2
V1=((8.98*(10**9))*(-1 * 10**-9)/((x+d)**2+y**2)**(1/2)) 
V2=((8.98*(10**9))*(-1 * 10**-9)/((x-d)**2+y**2)**(1/2)) 
V3=((8.98*(10**9))*(1 * 10**-9)/(x**2+(y+d)**2)**(1/2)) 
V4=((8.98*(10**9))*(1 * 10**-9)/(x**2+(y-d)**2)**(1/2)) 
V = V1 + V2 + V3 + V4
surf = ax.plot_surface(x, y, V, rstride=1, cstride=1, linewidth=0,cmap=cm.coolwarm)
xlabel("x")
ylabel("y")
```



(b) Now make a contour plot of the same data, with appropriate contour levels to see the shape of the potential.


```python
levels = array([-1, 0, 1])
plt.contour(x,y,V,levels,linewidths=4,cmap=cm.coolwarm);
```
